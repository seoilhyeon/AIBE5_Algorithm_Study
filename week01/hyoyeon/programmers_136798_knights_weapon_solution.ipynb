- 핵심: 약수는 짝으로 등장하므로 `sqrt(n)`까지만 확인한다.


## 문제 해석

각 기사 번호의 약수 개수가 필요한 공격력이다.  
하지만 약수 개수가 `limit`를 초과하면, 실제 개수 대신 `power`를 사용한다.  
결국 해야 할 일은 **1번부터 number번까지 각 수의 약수 개수**를 빠르게 구해서 합산하는 것이다.


## 처음 풀이가 느린 이유

가장 단순하게 풀면 각 수 `j`에 대해 `1 ~ j`를 전부 나눠 보게 된다.

```python
for j in range(1, number + 1):
    for i in range(1, j + 1):
        if j % i == 0:
            ...
```

이 방식은 반복 횟수가 너무 많아서 `number = 100000`일 때 시간 초과가 날 수 있다.


## 핵심 아이디어

### 1) 약수는 짝으로 등장한다
`i`가 `j`의 약수라면 `j // i`도 약수다.

예: 12의 약수 짝
- 1 x 12
- 2 x 6
- 3 x 4

따라서 `1 ~ j`를 전부 볼 필요 없이, `1 ~ sqrt(j)`까지만 보면 된다.

### 2) 제곱수는 한 번만 센다
`16`은 `4 x 4`처럼 가운데 약수가 겹친다.
그래서 `i == j // i`이면 `+1`, 아니면 `+2`를 해줘야 한다.

### 3) 실제 약수 목록은 필요 없다
이 문제는 약수 리스트를 저장할 필요가 없고, 개수만 세면 된다.


````python
def solution(number, limit, power):
    answer = 0

    for j in range(1, number + 1):
        total = 0

        for i in range(1, int(j ** 0.5) + 1):
            if j % i == 0:
                if j // i == i:
                    total += 1
                else:
                    total += 2

        if total > limit:
            answer += power
        else:
            answer += total

    return answer
````

````python
# 예제 확인
print(solution(5, 3, 2))   # 10
print(solution(10, 3, 2))  # 21
````

## 시간복잡도

각 수마다 `sqrt(n)`까지만 확인하므로 전체 시간복잡도는 대략 다음과 같다.

- `O(number * sqrt(number))`

완전탐색 `O(number^2)`보다 훨씬 효율적이다.


## 한 줄 정리

**약수는 짝으로 존재한다는 성질을 이용해서, 각 수의 약수 개수를 `sqrt(n)`까지만 세면 시간 초과 없이 해결할 수 있다.**
