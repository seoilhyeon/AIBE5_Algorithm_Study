# 프로그래머스 H-Index (Python)

정렬을 이용해 H-Index를 구하는 풀이를 정리한 노트북입니다.

## 문제 핵심

- 인용 횟수를 내림차순 정렬한다.
- `i`번째 논문이 `i+1`번 이상 인용되었는지 확인한다.
- 조건 `cit[i] >= i+1`를 만족하는 최대 개수가 H-Index이다.

```
def solution(citations):
    answer = 0
    n = len(citations)
    cit = sorted(citations, reverse=True)

    for i in range(n):
        if cit[i] >= i + 1:
            answer += 1

    return answer
```

## 예시 확인

입력 예시 `citations = [3, 0, 6, 1, 5]` 를 직접 넣어보면 정답은 3이다.

```
print(solution([3, 0, 6, 1, 5]))
```

## 시간복잡도

- 정렬: `O(n log n)`
- 순회: `O(n)`

따라서 전체 시간복잡도는 `O(n log n)`이다.

## 한 줄 정리

정렬 후 `cit[i] >= i+1`가 성립하는 최대 개수를 세면 된다.
