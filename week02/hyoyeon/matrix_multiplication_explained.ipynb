# 행렬의 곱 풀이 정리

이 노트북은 프로그래머스 `행렬의 곱셈` 문제를 기준으로,  
파이썬 풀이의 핵심 아이디어와 반복문 구조를 쉽게 이해할 수 있도록 정리한 자료입니다.

---

## 문제 핵심

두 행렬 `arr1`, `arr2`가 주어졌을 때,  
행렬 곱셈 결과를 2차원 리스트로 반환해야 합니다.

예를 들어:

- `arr1`의 크기: `m x n`
- `arr2`의 크기: `n x p`

라면 결과 행렬의 크기는:

- `m x p`

가 됩니다.

즉,

- 결과 행 개수 = `arr1`의 행 개수
- 결과 열 개수 = `arr2`의 열 개수

## 내가 푼 코드

```python
def solution(arr1, arr2):
    rows = len(arr1)
    cols = len(arr2[0])
    matrix = [[0] * cols for _ in range(rows)]

    for i in range(rows):
        for j in range(cols):
            val = 0
            for k in range(len(arr2)):
                val += arr1[i][k] * arr2[k][j]
            matrix[i][j] = val

    return matrix
```

## 왜 `matrix = [[0] * cols for _ in range(rows)]` 로 만드는가?

결과 행렬은 미리 `rows x cols` 크기의 빈 표를 만들어 두고,  
각 칸 `matrix[i][j]` 에 계산값을 넣는 방식으로 채웁니다.

예를 들어 결과가 `2행 3열`이라면 먼저 이런 틀을 만듭니다.

```python
[
    [0, 0, 0],
    [0, 0, 0]
]
```

여기서 `0`은 아직 계산 전이라는 뜻이고,  
나중에 `matrix[i][j] = val` 로 진짜 값을 넣게 됩니다.

## 반복문의 의미

행렬 곱셈은 보통 3중 반복문으로 풉니다.

### 1. `i`
결과 행렬의 **행 번호**

### 2. `j`
결과 행렬의 **열 번호**

### 3. `k`
곱해서 더해 줄 **연결 축**

즉 `matrix[i][j]` 는 아래 값들의 합입니다.

`arr1[i][0] * arr2[0][j] + arr1[i][1] * arr2[1][j] + ...`

그래서 코드가 이렇게 됩니다.

```python
for i in range(rows):
    for j in range(cols):
        val = 0
        for k in range(len(arr2)):
            val += arr1[i][k] * arr2[k][j]
        matrix[i][j] = val
```

## 한 칸이 계산되는 방식 예시

예를 들어:

```python
arr1 = [[1, 4],
        [3, 2]]

arr2 = [[3, 3],
        [3, 3]]
```

`matrix[0][1]` 을 구한다고 해 봅시다.

- `arr1[0]` 의 0번째 값과 `arr2[0][1]` 곱하기
- `arr1[0]` 의 1번째 값과 `arr2[1][1]` 곱하기
- 전부 더하기

즉,

`matrix[0][1] = 1 * 3 + 4 * 3 = 15`

이런 식으로 결과 행렬의 각 칸을 채우게 됩니다.

````python
def solution(arr1, arr2):
    rows = len(arr1)
    cols = len(arr2[0])
    matrix = [[0] * cols for _ in range(rows)]

    for i in range(rows):
        for j in range(cols):
            val = 0
            for k in range(len(arr2)):
                val += arr1[i][k] * arr2[k][j]
            matrix[i][j] = val

    return matrix

# 예시 실행
arr1 = [[1, 4], [3, 2], [4, 1]]
arr2 = [[3, 3], [3, 3]]

solution(arr1, arr2)
````

## 시간복잡도

반복문이 3중으로 돌기 때문에 시간복잡도는 다음과 같이 볼 수 있습니다.

- `rows = len(arr1)`
- `cols = len(arr2[0])`
- `shared = len(arr2)`

시간복잡도는 대략 `O(rows * cols * shared)` 입니다.

행렬 곱셈 문제에서는 아주 자연스러운 기본 풀이입니다.

## 포인트 한 줄 정리

이 문제는

- 결과 행렬의 크기를 먼저 만들고
- 결과의 각 칸 `matrix[i][j]` 마다
- `arr1`의 `i`행과 `arr2`의 `j`열을 곱해서 더한 값

을 넣는 문제입니다.
