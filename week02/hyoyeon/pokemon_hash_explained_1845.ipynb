# 프로그래머스 1845 폰켓몬

네가 푼 Python 풀이를 기준으로, 핵심 아이디어와 Java 변환까지 한 번에 정리한 노트북이다.


## 1. 문제 핵심

`nums`에는 폰켓몬 번호가 들어 있다.  
총 `N`마리 중에서 **정확히 `N/2`마리만 선택**할 수 있다.

목표는:

- 최대한 **많은 종류**의 폰켓몬을 선택하는 것

즉, 중요한 것은 각 번호가 **몇 번 나왔는지 자체**보다  
**서로 다른 종류가 몇 개인지**이다.


## 2. 왜 해시(Hash) 문제인가?

이 문제는 정답 계산 자체는 단순하지만, 먼저 해야 하는 일이 있다.

바로:

- 같은 번호끼리 묶기
- 서로 다른 종류의 개수 세기

이 작업을 가장 빠르게 해주는 자료구조가 **해시맵(dict)** 이다.

예를 들어:

```python
nums = [3, 1, 2, 3]
```

이면 종류는

- 1
- 2
- 3

총 3종류다.

이처럼 **값의 등장 여부 / 개수 / 종류 수**를 빠르게 관리하는 문제가 코딩테스트에서 흔히 '해시 문제'로 분류된다.


## 3. 핵심 아이디어

선택 가능한 마릿수는 무조건 `len(nums) // 2` 이다.

그리고 실제로 가져갈 수 있는 종류 수는 최대 두 가지 제한을 받는다.

1. **전체 종류 수**
2. **선택 가능한 마릿수**

예를 들어:

- 종류가 10개 있어도
- 내가 3마리만 고를 수 있으면

최대로 가져갈 수 있는 종류 수는 3개다.

반대로:

- 2종류밖에 없는데
- 5마리를 고를 수 있어도

최대로 가져갈 수 있는 종류 수는 2개다.

따라서 정답은:

```python
min(서로 다른 종류 수, len(nums)//2)
```

이 된다.


## 4. 네 Python 풀이

아래 풀이는 **dict로 종류를 세는 방식**이다.  
정답이며, 문제 의도에도 잘 맞는다.


````
def solution(nums):
    answer = 0
    dic = {}
    max_ans = int(len(nums)//2)

    for i in range(len(nums)):
        if nums[i] not in dic:
            dic[nums[i]] = 1
        else:
            dic[nums[i]] += 1

    count = 0
    for k, v in dic.items():
        count += 1

    if count <= max_ans:
        answer = count
    else:
        answer = max_ans
    return answer
````

## 5. 코드 흐름 설명

### (1) `max_ans = len(nums)//2`
내가 실제로 선택할 수 있는 최대 마릿수다.

예를 들어 `nums` 길이가 6이면  
정확히 3마리만 고를 수 있다.

---

### (2) `dic`에 번호별 개수 저장
```python
if nums[i] not in dic:
    dic[nums[i]] = 1
else:
    dic[nums[i]] += 1
```

이 부분은:

- 처음 본 번호면 새로 등록
- 이미 본 번호면 개수 증가

즉, 번호별 빈도를 세는 부분이다.

예:

```python
nums = [3, 1, 2, 3]
```

처리 후:

```python
dic = {3: 2, 1: 1, 2: 1}
```

여기서 중요한 것은 사실 빈도값보다  
**key가 3개 있다는 사실**, 즉 **종류가 3개**라는 점이다.

---

### (3) `count`로 종류 수 세기
```python
for k, v in dic.items():
    count += 1
```

dict에 들어 있는 key 개수만큼 종류 수가 존재한다는 뜻이다.

즉:

- `dic`의 key 개수 = 서로 다른 폰켓몬 종류 수

---

### (4) 종류 수와 고를 수 있는 수 비교
```python
if count <= max_ans:
    answer = count
else:
    answer = max_ans
```

- 종류 수가 더 적으면 종류 수만큼만 가능
- 고를 수 있는 마릿수가 더 적으면 그 수만큼만 가능

그래서 결국 `min(count, max_ans)` 와 같은 의미다.


## 6. 더 짧게 쓰면?

네 풀이는 정답이다.  
다만 Python에서는 보통 아래처럼 더 짧게 쓸 수 있다.

### dict 버전 간단화
```python
def solution(nums):
    dic = {}
    for n in nums:
        if n not in dic:
            dic[n] = 1
        else:
            dic[n] += 1
    return min(len(dic), len(nums)//2)
```

### set 버전
```python
def solution(nums):
    return min(len(set(nums)), len(nums)//2)
```

`set(nums)` 는 중복을 제거하므로  
곧바로 서로 다른 종류 수를 구할 수 있다.


````
def solution_dict(nums):
    dic = {}
    for n in nums:
        if n not in dic:
            dic[n] = 1
        else:
            dic[n] += 1
    return min(len(dic), len(nums)//2)

def solution_set(nums):
    return min(len(set(nums)), len(nums)//2)

print(solution_dict([3, 1, 2, 3]))  # 2
print(solution_set([3, 1, 2, 3]))   # 2
````

## 8. 시간복잡도

- dict 또는 set를 이용해 한 번 순회하므로
- 시간복잡도는 **O(N)**

코딩테스트에서 매우 효율적인 풀이이다.


## 9. Java로 변환하면

Java에서는 `HashMap<Integer, Integer>` 를 사용해 같은 방식으로 풀 수 있다.


````
import java.util.HashMap;
import java.util.Map;

class Solution {
    public int solution(int[] nums) {
        Map<Integer, Integer> map = new HashMap<>();

        // 1. 번호별 개수 세기
        for (int num : nums) {
            map.put(num, map.getOrDefault(num, 0) + 1);
        }

        // 2. 서로 다른 종류 수
        int kindCount = map.size();

        // 3. 선택 가능한 최대 마릿수
        int maxPick = nums.length / 2;

        // 4. 둘 중 작은 값이 정답
        return Math.min(kindCount, maxPick);
    }
}
````

## 10. Java 코드 설명

### `Map<Integer, Integer> map = new HashMap<>();`
- key: 폰켓몬 번호
- value: 해당 번호가 나온 횟수

### `map.put(num, map.getOrDefault(num, 0) + 1);`
- 처음 나온 번호면 0에서 시작
- 이미 있으면 기존 값 + 1

### `map.size()`
- 서로 다른 종류 수

### `nums.length / 2`
- 내가 가져갈 수 있는 최대 마릿수

### `Math.min(kindCount, maxPick)`
- 실제 정답


## 11. 한 줄 정리

이 문제의 본질은:

- **서로 다른 종류 수를 구하고**
- **고를 수 있는 최대 수 `N/2` 와 비교해서**
- **더 작은 값을 반환하는 문제**

즉:

```python
정답 = min(서로 다른 종류 수, len(nums)//2)
```
